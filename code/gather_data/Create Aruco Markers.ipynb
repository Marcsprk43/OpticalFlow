# Create a set of aruco markers
This notebook creates a set of aruco markers for the experimental targets

```python
import cv2
from cv2 import aruco
import matplotlib.pyplot as plt
```

```python
aruco_dict = aruco.Dictionary_get(aruco.DICT_4X4_50)
```

```python
workdir = "/Users/gerrie/Google Drive/ScienceFair/aruco markers/"

marker_number = 0
for marker_number in range(30):

    marker =  aruco.drawMarker(aruco_dict, marker_number, 1000, 1)

    cv2.imwrite(workdir + "ArucoMarker_DICT_4X4_50_{}.tiff".format(marker_number), marker)


    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    plt.imshow(marker, cmap = plt.cm.gray, interpolation = "nearest")
    ax.axis("off")
    plt.show()
```
